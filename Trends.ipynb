**Determining the maxima and minima of each trend term**

Create vector of year and week numbers

```python
dates = trends.week_start
yrs = zeros_like(dates)
wks = zeros_like(dates)
for i in range(len(dates)):
    yrs[i] = dates[i].year
    wks[i] = dates[i].isocalendar()[1]
```

For each year list week numbers corresponding to maximum and minimum search values 

```python
trend = trends.global_warming
for yr in range(2004,2011):
    index = find(yrs==yr)
    print yr, wks[find(trend[index] == max(trend[index]))], wks[find(trend[index] == min(trend[index]))]
```

**Study scatter about the median values**

We want to find which search term has the largest scatter over time. The question is: Is it true that $\sigma(A) > \sigma(B) \Rightarrow \sigma_\textrm{median}(A) > \sigma_\textrm{median}(B)$?

```python
def std_median(datums):
    return np.sqrt( np.sum( (datums - datums.median())**2 ) * 1.0/ len(datums) )
```

```python
print "Spring break: ", df.spring_break.std(), std_median(df.spring_break)
print "Textbooks: ", df.textbooks.std(), std_median(df.textbooks)
print "Skiing: ", df.skiing.std(), std_median(df.skiing)
print "Kayak: ", df.kayak.std(), std_median(df.kayak)
print "Global Warming: ", df.global_warming.std(), std_median(df.global_warming)
```

We see that searches for skiing have the highest scatter, whereas those for kayak remain more stable. 

**Finding the correlation between trends**

We can use cross-correlation to find the time delay between two time-series. The maximum of the cross-correlation function gives the time when the two signals are best aligned (in other words, the lag between the two signals).

```python
result = np.correlate(df['spring_break'],df['skiing'], mode='full')
gap = np.arange(result.size) - result.size/2
plot(gap,result)
```

We can use the numpy function argmax to get the index at which this value is maximum (note for pandas expert - result appears still to be a pandas object and maintains attributes such as idxmax, but they don't work?)

```python
print gap[result.argmax()] # Find the x value corresponding to the maximum y value
```

```python
result2 = np.correlate(df.global_warming,df.skiing, mode='full')
gap2 = np.arange(result2.size) - result2.size/2
plot(gap2,result2)
```

```python
print gap2[result2.argmax()]  # Find the x value corresponding to the maximum y value
```
